```python
import PIL
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img_path = 'noidea.jpg'
```

```python
display(img_path)
```

```python
image = Image.open(img_path)

pixel_matrix = image.load()
```

```python
display(image)
```

```python
#width and height of the image
width, height = image.size
print(width,height)
```

```python
## RGB TO GrayScale

#  r+g+b/3

for i in range(width):
  for j in range(height):

    r,g,b = image.getpixel((i,j))

    grayscale = (r+g+b)/3

    #setting the pixel value
    pixel_matrix[i,j] = (int(grayscale), int(grayscale), int(grayscale))
```

```python
display(image)
```

```python
#for better human eye view there is another formula instead of using mean while converting from rgb to gscale
## RGB TO GrayScale

#  (0.299R + 0.587G + 0.114B)

for i in range(width):
  for j in range(height):

    r,g,b = image.getpixel((i,j))

    grayscale = (0.299*r + 0.587*b + 0.114*b)

    #setting the pixel value
    pixel_matrix[i,j] = (int(grayscale), int(grayscale), int(grayscale))
```

```python
image.save('grayscale.png',format="png")  #saving the final output image
```

```python
display(image)
```
